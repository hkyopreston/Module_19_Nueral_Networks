```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
pd.options.display.max_columns = None
df = pd.read_csv('../Resources/AmesHousing.csv')
### metadata: https://rdrr.io/cran/AmesHousing/man/ames_raw.html
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df.dtypes)
```

```python
numeric_columns = ['Lot Frontage','Lot Area', 'Overall Qual', 'Overall Cond', 'Year Built', 'Year Remod/Add', 'Mas Vnr Area', 
                   'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', '1st Flr SF', '2nd Flr SF', 
                   'Low Qual Fin SF', 'Gr Liv Area', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Full Bath', 'Half Bath', 
                   'Bedroom AbvGr', 'Kitchen AbvGr', 'TotRms AbvGrd', 'Fireplaces', 'Garage Yr Blt', 'Garage Cars', 
                   'Garage Area', 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 'Screen Porch', 'Pool Area', 
                   'Misc Val', 'Mo Sold', 'Yr Sold']
category_columns = ['MS SubClass', 'MS Zoning', 'Street', 'Alley', 'Lot Shape', 'Land Contour', 'Utilities', 'Lot Config', 
                    'Land Slope', 'Neighborhood', 'Condition 1', 'Condition 2', 'Bldg Type', 'House Style', 'Roof Style', 
                    'Roof Matl', 'Exterior 1st', 'Exterior 2nd', 'Mas Vnr Type', 'Mas Vnr Area', 'Exter Qual', 'Exter Cond', 
                    'Foundation', 'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2', 'Heating', 
                    'Heating QC', 'Central Air', 'Electrical', 'Kitchen Qual', 'Functional', 'Fireplace Qu', 'Garage Type', 
                    'Garage Finish', 'Garage Qual', 'Garage Cond', 'Paved Drive', 'Pool QC', 'Fence', 'Misc Feature', 
                    'Sale Type', 'Sale Condition']
```

```python
X = df[numeric_columns].merge(pd.get_dummies(df[category_columns]), left_index=True, right_index=True)
X
```

```python
X = X.fillna(0)
y = df['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
nn_model = tf.keras.models.Sequential()
```

```python
nn_model.add(tf.keras.layers.Dense(units=len(X.columns) * 2, activation = "relu", input_dim = len(X.columns)))
```

```python
nn_model.add(tf.keras.layers.Dense(units=1, activation="linear"))
```

```python
nn_model.summary()
```

```python
nn_model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
nn_model.fit(X_train_scaled, np.asarray(y_train), epochs=100)
```

```python
y_train_pred = nn_model.predict(X_train_scaled)
y_test_pred = nn_model.predict(X_test_scaled)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

```python
deep_model = tf.keras.models.Sequential()
```

```python
deep_model.add(tf.keras.layers.Dense(units=len(X.columns) * 2, activation = "relu", input_dim = len(X.columns)))
```

```python
deep_model.add(tf.keras.layers.Dense(units=len(X.columns) * 2, activation = "relu"))
```

```python
deep_model.add(tf.keras.layers.Dense(units=1, activation="linear"))
```

```python
deep_model.summary()
```

```python
deep_model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
deep_model.fit(X_train_scaled, np.asarray(y_train), epochs=100)
```

```python
y_train_pred = deep_model.predict(X_train_scaled)
y_test_pred = deep_model.predict(X_test_scaled)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```
